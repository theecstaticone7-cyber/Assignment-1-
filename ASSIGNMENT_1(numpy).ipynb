#####Problem:3
#####Find the most frequent value and their indice(s) in the following arrays:
#####i. x = np.array([1,2,3,4,5,1,2,1,1,1])
#####ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])

def freq_and_index(a):
    m = int(np.bincount(a).argmax())
    return m, np.where(a == m)[0]

print(freq_and_index(x))
print(freq_and_index(y))
```

#####Problem:4
#####For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
#####find i. Sum of all elements
#####ii. Sum of all element’s row-wise
#####iii. Sum of all element’s column-wise

```python
gfg = np.matrix('4 1 9; 12 3 1; 4 5 6')
print('sum of all values: ')
print(np.sum(gfg))

print('sum of all rows: ')
print(np.sum(gfg, axis=1))

print('sum of all columns: ')
print(np.sum(gfg, axis=0))
```
