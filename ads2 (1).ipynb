3. Debug common CUDA errors (zero output, incorrect indexing, PTX errors)


```python
# Zero Output Error:
#Without synchronization, GPU output may not appear
%%writefile hi.cu

#include <stdio.h>

__global__ void zeroOutputKernel() {
    printf("Hello from CUDA\n");
}

int main() {
    zeroOutputKernel<<<1,1>>>();
    cudaDeviceSynchronize();   // Fix: synchronize GPU
    return 0;
}
```

```python
# Incorrect Indexing Error:
# Correct global thread ID calculation
%%writefile hi.cu

#include <stdio.h>

__global__ void indexingKernel() {
    int global_id = blockIdx.x * blockDim.x + threadIdx.x;
    printf("Global Thread ID: %d\n", global_id);
}

int main() {
    indexingKernel<<<2,4>>>();   // 2 blocks, 4threads
    cudaDeviceSynchronize();
    return 0;
}
```

Write a CUDA C/C++ program to demonstrate GPU kernel execu'on and thread indexing.

a. Launch a CUDA kernel using: 1 block and 8 threads

b. Each thread must print: Hello from GPU thread <global_thread_id>

c. Compute the global thread ID using: global_thread_id = blockIdx.x * blockDim.
x + threadIdx.x

d. Clearly separate: Host code (CPU) & Device code (GPU kernel)

```python
%%writefile thread_index.cu
// Demonstration of GPU kernel execution and thread indexing

#include <stdio.h>

//  DEVICE CODE (GPU Kernel)
__global__ void helloKernel() {
    int global_thread_id = blockIdx.x * blockDim.x + threadIdx.x;
    printf("Hello from GPU thread %d\n", global_thread_id);
}

// HOST CODE (CPU)
int main() {

    helloKernel<<<1, 8>>>();
    cudaDeviceSynchronize();

    return 0;
}
```

```python
# Compile the CUDA program
!nvcc -arch=sm_75 thread_index.cu -o thread_index
```

```python
# Execute the CUDA program
!./thread_index
```

Write a CUDA program to demonstrate host and device memory separation.

a. Create an integer array of size 5 on the host (CPU).

b. Allocate corresponding memory on the device (GPU) using cudaMalloc().

c. Copy data from host to device using cudaMemcpy().

d. Launch a kernel where GPU threads print values from device memory.

e. Copy the data back from device to host and print it on CPU.

```python
%%writefile device_memory.cu
#include <stdio.h>
#include <cuda_runtime.h>

//  DEVICE CODE
__global__ void printDeviceArray(int *d_arr) {
    int id = threadIdx.x;
    printf("GPU Thread %d : Value = %d\n", id, d_arr[id]);
}

//  HOST CODE
int main() {

    // (a) Create integer array on host (CPU)
    int h_arr[5] = {10, 20, 30, 40, 50};

    int *d_arr;

    // (b) Allocate memory on device (GPU)
    cudaMalloc((void**)&d_arr, 5 * sizeof(int));

    // (c) Copy data from host to device
    cudaMemcpy(d_arr, h_arr, 5*sizeof(int), cudaMemcpyHostToDevice);

    // (d) Launch kernel to print values from device memory
    printDeviceArray<<<1,5>>>(d_arr);
    cudaDeviceSynchronize();

    // (e) Copy data back from device to host
    cudaMemcpy(h_arr, d_arr, 5 * sizeof(int), cudaMemcpyDeviceToHost);
    printf("\nValues copied back to Host (CPU):\n");
    for (int i = 0; i < 5; i++) {
        printf("Host index %d : %d\n", i, h_arr[i]);
    }
    cudaFree(d_arr);

    return 0;
}
```

```python
!nvcc -arch=sm_75 device_memory.cu -o dm
```

```python
!./dm
```
